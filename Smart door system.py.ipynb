```python
class Staff:
    def __init__(self, s_name, access_code):
        self.s_name = s_name
        self.__access_code = access_code

    # Property getter
    @property
    def access_code(self):
        print("Authorized access granted.")
        return self.__access_code

    # Property setter
    @access_code.setter
    def access_code(self, new_code):
        if len(new_code) >= 6:
            self.__access_code = new_code
            print("Access code updated successfully.")
        else:
            print("Error: Access code must be at least 6 characters long.")

    # Display staff information
    def display_info(self):
        print("\n===== Staff Information =====")
        print(f"Staff Name : {self.s_name}")
        print(f"Access Code: {self.__access_code}")


# -----------------------------
# Testing the program
# -----------------------------
staff1 = Staff("Yaw Solomon", "UMA12345")

# Display information
staff1.display_info()

# View access code
print("\nViewing Access Code:")
print(staff1.access_code)

# Update with an invalid code
print("\nUpdating Access Code with '123'")
staff1.access_code = "123"

# Update with a valid code
print("\nUpdating Access Code with 'ORIC2026'")
staff1.access_code = "ORIC2026"

# Display updated information
staff1.display_info()
```
